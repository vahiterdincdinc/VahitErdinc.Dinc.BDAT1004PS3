```python
import random
first=[]
second=[]
third=[]
for i in range(100):
    first.append(random.randrange(1,4))
    second.append(random.randrange(1,3))
    third.append(random.randrange(10000,30000))
firstCol=pd.Series(first,name="FirstCol")
secondCol=pd.Series(second,name="SecondCol")
thirdCol=pd.Series(third,name="ThirdCol")
```

## Step 3

```python
data=pd.concat([firstCol,secondCol,thirdCol],axis=1)
data
```

## Step 4

```python
data.rename(columns={'FirstCol':'bedrs', 'SecondCol':'bathrs', 'ThirdCol':'price_sqr_meter'},inplace=True)
data
```

```python
import matplotlib.pyplot as plt
from matplotlib import figure
import sys
import pandas as pd
data=pd.read_csv("us-marriages-divorces-1867-2014.csv")
data
```

```python
plt.figure(figsize=(10,8))
plt.title('Number of Marriges and Divorces', fontdict={'fontsize': 15})
plt.plot(data.Year, data.Marriages_per_1000, 'y.-', label='Marriage')
plt.plot(data.Year, data.Divorces_per_1000, 'r.-', label='Divorce')
plt.xlabel('Through Years')
plt.ylabel('Marriages and Divorces per Capitas')
plt.legend()
plt.show()
```

```python
import pandas as pd
import seaborn as sns
deadliest_actors = pd.read_csv('actor_kill_counts.csv')
res1 = sns.barplot(x='Count', y='Actor', data=deadliest_actors, order=deadliest_actors.sort_values('Count', ascending=True).Actor)
res1.set_xlabel('Total Kills Number')
res1.set_ylabel('Actor Names')
res1.set_title("Hollywood's Deadliest actors ", size=17)
```
